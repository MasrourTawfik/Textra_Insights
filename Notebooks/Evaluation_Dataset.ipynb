### Run On Selected Invoices

````python
import time
import os
from tqdm import tqdm
import pandas as pd

################################################################################
################################### PaddleOCR ##################################
from paddleocr import PaddleOCR
paddle_reader = PaddleOCR(use_angle_cls=True, lang='fr')

def Run_Paddle(Image_path):
    Start = time.time()
    result = paddle_reader.ocr(Image_path, det=True, rec=True)
    End = time.time()
    # Inference Time
    Time = End-Start
    # Process the result
    extracted_text = ''
    #print("Lenght Result: ",len(result))
    for line in result:
        if line is not None:
            for word in line:
                #print("Word: ",word)
                # Access the text part of the tuple
                extracted_text += word[1][0] + ' '  # Access the first element of the recognized text (the text itself)
            extracted_text += '\n'
    return extracted_text,Time


def Run_Paddle_ALL(Folder_path):
    df = pd.DataFrame(columns=['InvoiceName', 'Text', 'Time(s)'])
    Files = os.listdir(Folder_path)
    Files = sorted(Files, key=lambda x: int(x.split('.')[0]))
    Texts = []
    Times = []
    for file in tqdm(Files,desc="Paddle Processing", unit="number"):
        print(f"{file} is processing")
        Invoice_path = os.path.join(Folder_path, file)
        Text_paddle,Time_paddle = Run_Paddle(Invoice_path)
        Texts.append(Text_paddle)
        Times.append(Time_paddle)
    df['InvoiceName'] = Files
    df['Text'] = Texts
    df['Time(s)'] = Times

    # Saving the DataFrame to a CSV file
    df.to_csv('PaddleOCR_Results.csv', index=False)
````

````python
Run_Paddle_ALL(New_Folder_Path)
````

````python
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field
from langchain_openai import ChatOpenAI

class Description(BaseModel):
    analyse: str = Field("", description="Analyse détaillée de la transaction.")
    proposition: str = Field("", description="Intitulé du compte à débiter dans cette facture.")
    nature : str = Field("", description="Nature du produit ou service acheté (ex. : 'Matériel informatique', 'Fournitures de bureau', etc.).")
    Keywords : str = Field("", description="Les mots-clés dans la facture liés à une opération d'achat. Par exemple : Ordinateur, Service, Machine, Eau, Énergie...")


system = f"""

*Vous êtes un expert comptable chargé d'analyser une facture d'achat en vous basant sur le texte extrait de cette facture. Votre tâche consiste à extraire et à classifier les informations suivantes :*

1. **Analyse détaillée de la transaction** :

    - Fournissez une analyse comptable détaillée de la transaction.

2. **Proposition du compte à débiter** :

   - Sur la base de l'analyse de la transaction, proposez un **compte à débiter** approprié pour cette facture (par exemple, **Fournitures de bureau** ou **Achats de combustibles**).

3. **Nature du produit ou service acheté** :
   - Précisez la **nature spécifique** du produit ou service acheté, comme par exemple **Matériel informatique**, **Fournitures de bureau**, **Prestations de services**, etc.

4. **Extraction des mots-clés** :
  - Identifiez les mots-clés directement liés à l'achat mentionnés dans la facture. Ces mots-clés doivent correspondre aux éléments caractéristiques de l'achat (exemple : Ordinateur, Service, Machine, Eau, Énergie, etc.). Assurez-vous que ces mots-clés sont représentatifs et facilitent la classification des informations.


### **Exemple d'entrée** :

```
Achat de 10 ordinateurs portables et 2 imprimantes pour l'entreprise.

```

### **Exemple de sortie attendue** :

```
analyse: "L'achat concerne du matériel informatique destiné à être utilisé de manière durable dans les opérations de l'entreprise. Il s'agit donc d'une immobilisation."
proposition: "215 - Matériel informatique"
nature: "Matériel informatique"
Keywords: "Ordinateur, Imprimante"
```
"""

## Prompt Template
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system),
        ("human", """
        Texte extrait à partir de la facture ici :\n {OCR_Output} \n
        Quelle est votre analyse de cette facture ?
        """),
    ]
)

LLM = ChatOpenAI(model="llama3.1:8b",
    #temperature=0,
    api_key='ollama',
    base_url="http://localhost:11434/v1",
)
LLM_Structered = LLM.with_structured_output(Description)
LLM_descriptor = prompt | LLM_Structered
````

````python
def Run_LLM(OCR_Output):
    Input = {"OCR_Output":OCR_Output}
    response = LLM_descriptor.invoke(Input)
    return response
````

````python
import pandas as pd

OCR_Results = pd.read_csv('PaddleOCR_Results.csv')
Invoice_Text = OCR_Results['Text'][5]
Invoice_Text
````

## Why Raw Definitions Are Bad ?

````python
Definition1 = """
6125. Achats non stockés de matières et de fournitures :

Ce compte enregistre les achats de matières et fournitures non stockables tels que l’eau,
l’électricité et autres énergies, ainsi que les achats de matières et fournitures non stockées par
l’entreprise (les matières et fournitures dont la gestion des stocks est jugée inutile par l’entreprise).
"""

Definition2 = """
6126. Achats de travaux, études et prestations de services :

Ce compte enregistre les divers travaux et services d'exploitation confiés, en partie, par l'entreprise à autre personne.
"""

Definition3 = """
6145. Frais postaux et frais de télécommunication :

Ce compte enregistre les frais postaux (achats de timbres, achats d’enveloppes préaffranchies ...),
les frais de téléphone, les frais de transmission de messages dactylographiques (service de télex) et
les frais de télégramme (message transmis par voie électrique ou radioélectrique). Cependant,
les frais engagés pour l’envoi des fax sont aussi à comptabiliser dans ce compte puisqu’ils
font aussi partie des frais de télécommunication.
"""
````

````python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Charger un modèle pré-entraîné de Sentence Transformers
model = SentenceTransformer('xlm-r-base-en-ko-nli-ststb')
````

````python
# Liste des définitions
definitions = [
    Definition1,
    Definition2,
    Definition3
]


# Encoder la chaîne de réponse une seule fois
embeddings2 = model.encode([Query])

# Itérer sur les définitions et calculer les similarités
for i, definition in enumerate(definitions):
    embeddings1 = model.encode([definition])  # Encoder chaque définition
    similarity = cosine_similarity(embeddings1, embeddings2)  # Calculer la similarité
    print(f"Score de similarité pour la définition {i + 1} : {similarity[0][0]}")
````
